## Ejercicio 1 - Gráfico de líneas simple

Supongamos que, en tu oficina, te toma 10 segundos recorrer 25 metros caminando tranquilamente hacia la máquina de café. Si queremos expresar esta situación como una función lineal de la distancia respecto al tiempo, usamos la fórmula:

$
\text{distancia} = \text{velocidad} \times \text{tiempo}
$



Dado que la velocidad es constante, esta representa la pendiente de la función y se puede interpretar como la primera derivada de la distancia con respecto al tiempo. Es decir, nos indica qué tan rápido cambia la distancia mientras pasa el tiempo.

- **Calcula la velocidad constante a la que caminas:** Sabemos que recorriste 25 metros en 10 segundos. ¿Cuál es la velocidad (en metros/segundo)?
- **Construye la función lineal que describe cómo cambia la distancia con el tiempo:** Usa la fórmula `distancia = velocidad × tiempo` para calcular la distancia recorrida en cada segundo desde 0 a 10.
- **Usa Matplotlib para graficar cómo cambia la distancia con el tiempo:** Agrega etiquetas a los ejes y un título al gráfico.

````python
x = np.arange(0, 11) # array con valores de 0 a 10

# Cálculo de la velocidad constante
velocity = 25 / 10

# Construcción de la función: distancia = velocidad * tiempo
y = velocity * x

# Visualización con Matplotlib
plt.figure(figsize=(7, 4))
plt.plot(x, y, label='Trayectoria al café')
plt.title('Distancia vs Tiempo (v = 2.5 m/s)')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Distancia (metros)')
plt.grid(True)
plt.legend()
plt.show()

````

### Notas de Resolución y Teoría a Recordar
#### Histograma de frecuencia y curva KDE
En el código mostrado se utiliza:
`sns.histplot(data=df_iris, x='petal_length', bins=20)`
Esto genera un **histograma de frecuencias** de la variable `petal_length`.
#### ¿Qué es un histograma de frecuencia?
Un **histograma** es una representación gráfica que muestra **cómo se distribuyen los valores de una variable numérica**.
Divide el rango de valores en intervalos llamados **bins** (contenedores o clases).
Cada barra representa:
- un intervalo de valores
- la **frecuencia** de datos dentro de ese intervalo
En este caso:
`bins=20`
significa que el rango de `petal_length` se divide en **20 intervalos**.
#### Información que nos proporciona un histograma
Un histograma permite analizar varias propiedades importantes de los datos.
#### 1. Distribución de los valores
Permite ver **cómo se reparten los datos** dentro del rango de la variable.
Ejemplo:
- si la mayoría de valores se concentran en un rango concreto
- si los datos están dispersos
#### 2. Forma de la distribución
La forma puede indicar diferentes comportamientos estadísticos:
**Distribución normal (campana)**  
Los datos se concentran en el centro y disminuyen hacia los extremos.
**Distribución sesgada (skewed)**  
Los datos se inclinan hacia un lado:
- sesgo positivo → cola hacia la derecha
- sesgo negativo → cola hacia la izquierda
#### 3. Concentraciones de datos (modos)
El histograma permite identificar **modos**.
Un **modo** es una zona donde se concentran muchos valores.
Ejemplos:
- **unimodal** → un solo pico
- **bimodal** → dos picos
- **multimodal** → varios picos
En el dataset **Iris**, `petal_length` suele mostrar **varios picos** porque las especies tienen tamaños de pétalo distintos.
#### 4. Dispersión de los datos
El histograma permite observar:
- si los datos están muy concentrados
- si están muy dispersos
Esto da una idea visual de la **variabilidad de la variable**.
#### 5. Posibles valores atípicos (outliers)
Valores muy alejados del resto pueden aparecer como **barras aisladas en los extremos** del histograma.
#### ¿Qué es KDE?
**KDE** significa:
**Kernel Density Estimation**
En español:
**Estimación de densidad por núcleo**
Es una técnica que **estima la distribución de probabilidad continua** de una variable.
En lugar de mostrar barras discretas como el histograma, KDE dibuja **una curva suave** que representa la densidad de los datos.
#### ¿Qué información aporta KDE?
La curva KDE permite observar con mayor claridad:
#### 1. Forma real de la distribución
El histograma depende de los **bins**, lo que puede alterar la apariencia.
KDE elimina ese problema y muestra una **curva suave de densidad**.
#### 2. Número de picos de la distribución
La curva KDE permite identificar fácilmente:
- uno o varios picos
- subpoblaciones dentro de los datos
Esto es muy útil en datasets como **Iris**, donde distintas especies generan picos diferentes.
#### 3. Tendencia de concentración de los datos
Las zonas más altas de la curva indican **mayor densidad de valores**.

Sin embargo, existe **solapamiento entre especies**, por lo que esta variable **no es suficiente por sí sola para clasificar perfectamente las flores**.


Personalmente me pareció bastante complicado entender los solapamientos aquí, así que, he modificado esta representación para que sea mas específica, como a mí me hubiera gustado recibirla.

````python

from matplotlib.patches import Patch

# --------------------------------------------------
# 1. Preparar nombres en español
# --------------------------------------------------
df_plot = df_iris.copy()

mapa_especies = {
    'setosa': 'Setosa',
    'versicolor': 'Versicolor',
    'virginica': 'Virginica'
}

df_plot['Especie_es'] = df_plot['species'].map(mapa_especies)

# --------------------------------------------------
# 2. Definir bins comunes para todas las especies
# --------------------------------------------------
bins = np.linspace(df_plot['petal_length'].min(), df_plot['petal_length'].max(), 16)
bin_centers = (bins[:-1] + bins[1:]) / 2
bin_width = np.diff(bins)

# --------------------------------------------------
# 3. Contar frecuencias por especie en cada bin
# --------------------------------------------------
c_setosa, _ = np.histogram(
    df_plot.loc[df_plot['Especie_es'] == 'Setosa', 'petal_length'],
    bins=bins
)

c_versicolor, _ = np.histogram(
    df_plot.loc[df_plot['Especie_es'] == 'Versicolor', 'petal_length'],
    bins=bins
)

c_virginica, _ = np.histogram(
    df_plot.loc[df_plot['Especie_es'] == 'Virginica', 'petal_length'],
    bins=bins
)

# --------------------------------------------------
# 4. Descomposición del solapamiento por bin
#    La idea es separar:
#    - solapamiento triple
#    - solapamientos por pareja
#    - partes exclusivas
# --------------------------------------------------
triple = np.minimum.reduce([c_setosa, c_versicolor, c_virginica])

r_setosa = c_setosa - triple
r_versicolor = c_versicolor - triple
r_virginica = c_virginica - triple

sv = np.minimum(r_setosa, r_versicolor)
r_setosa -= sv
r_versicolor -= sv

sg = np.minimum(r_setosa, r_virginica)
r_setosa -= sg
r_virginica -= sg

vg = np.minimum(r_versicolor, r_virginica)
r_versicolor -= vg
r_virginica -= vg

solo_setosa = r_setosa
solo_versicolor = r_versicolor
solo_virginica = r_virginica

# --------------------------------------------------
# 5. DataFrame auxiliar para inspección si lo necesitas
# --------------------------------------------------
df_overlap = pd.DataFrame({
    'bin_left': bins[:-1],
    'bin_right': bins[1:],
    'Setosa': solo_setosa,
    'Versicolor': solo_versicolor,
    'Virginica': solo_virginica,
    'Setosa + Versicolor': sv,
    'Setosa + Virginica': sg,
    'Versicolor + Virginica': vg,
    'Setosa + Versicolor + Virginica': triple
})

# --------------------------------------------------
# 6. Colores pastel + tramas para máxima claridad
# --------------------------------------------------
styles = [
    ('Setosa', '#b3d9ff', None),
    ('Versicolor', '#b8e6b8', None),
    ('Virginica', '#ffd1a9', None),
    ('Setosa + Versicolor', '#9fd3c7', '//'),
    ('Setosa + Virginica', '#d6c2ff', '\\\\'),
    ('Versicolor + Virginica', '#ffe599', 'xx'),
    ('Setosa + Versicolor + Virginica', '#cfcfcf', '++')
]

# --------------------------------------------------
# 7. Dibujar barras apiladas
# --------------------------------------------------
plt.figure(figsize=(12, 6))
bottom = np.zeros_like(bin_centers, dtype=float)

for label, color, hatch in styles:
    values = df_overlap[label].values
    plt.bar(
        bin_centers,
        values,
        width=bin_width * 0.95,
        bottom=bottom,
        color=color,
        edgecolor='black',
        linewidth=0.8,
        hatch=hatch,
        align='center',
        label=label
    )
    bottom += values

plt.title('Distribución de Largo del pétalo con solapamientos explícitos por especie')
plt.xlabel('Largo del pétalo')
plt.ylabel('Frecuencia por intervalo')
plt.legend(title='Especies / Solapamientos', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid(axis='y', alpha=0.25)
plt.tight_layout()
plt.show()
````

### Interpretación del histograma con solapamientos explícitos
Este gráfico representa la **distribución del largo del pétalo (`Largo del pétalo`) por especie**, pero a diferencia de un histograma tradicional, descompone cada intervalo en **partes exclusivas y partes de solapamiento entre especies**. Para ello se utilizan **colores pastel para las zonas exclusivas** y **tramas (texturas)** para indicar combinaciones de especies que aparecen simultáneamente en el mismo rango de valores.
    De esta forma, la leyenda no solo identifica cada especie, sino también los posibles solapamientos: `Setosa + Versicolor`, `Setosa + Virginica`, `Versicolor + Virginica` y `Setosa + Versicolor + Virginica`.
El objetivo de este gráfico es **hacer visible en qué rangos de la variable las especies se diferencian bien y en cuáles se mezclan**. Cada barra corresponde a un intervalo de `Largo del pétalo`, y su altura representa la frecuencia total de flores en ese intervalo, mientras que las subdivisiones muestran qué especies contribuyen a esa frecuencia. Cuando aparece una zona con trama significa que **más de una especie tiene observaciones dentro de ese mismo rango de valores**, lo que indica que la variable no separa completamente esas clases.

Se interpretaría este gráfico buscando **zonas de separación y zonas de solapamiento entre distribuciones**. Si una especie domina claramente un rango del eje X, esa variable tiene **alto poder discriminativo** para esa clase. En cambio, si aparecen muchas áreas con tramas de dos o tres especies, significa que **la variable por sí sola no permite distinguir bien entre ellas** y que será necesario utilizar otras variables del dataset o un modelo multivariable para clasificar correctamente las observaciones.
Aplicando esto a nuestros datos concluiríamos:
- Se observa cómo se distribuyen los valores de **`Largo del pétalo`** aproximadamente entre **1 y 8 unidades** y cómo se solapan las especies dentro de los distintos intervalos del histograma.
- En los valores **más bajos (≈1–2)** aparecen principalmente observaciones de **versicolor**, junto con algunos **solapamientos con virginica**.
- Entre **≈2 y 4** comienza una zona con bastante presencia de **las tres especies simultáneamente**, lo que se refleja en las áreas con la trama correspondiente a **`Setosa + Versicolor + Virginica`**.
- En la **zona central del gráfico (≈3–5)** es donde se observa el **mayor nivel de mezcla entre especies**, indicando que en ese rango de longitud de pétalo la variable **no permite distinguir claramente entre clases**.
- A partir de **≈5–6** aparecen más frecuencias asociadas a **virginica y versicolor**, incluyendo solapamientos entre ambas (**`Versicolor + Virginica`**).
- En los valores **más altos (≈7–8)** las observaciones son **escasas** y se concentran principalmente en **versicolor o virginica**.

En conjunto, el gráfico muestra que **la variable `Largo del pétalo` contiene información útil para diferenciar especies**, ya que los valores extremos tienden a asociarse con especies concretas. Sin embargo, existe un **amplio solapamiento en la zona media (≈3–5)**, lo que indica que esta variable por sí sola **no separa completamente las especies**, por lo que sería necesario considerar **otras variables del dataset** para realizar una clasificación más precisa.


### Por qué en el último gráfico no se utilizó `hue`

En los gráficos anteriores se utilizó el argumento:
`hue='species'`
El parámetro **`hue`** en Seaborn sirve para **separar automáticamente los datos según una variable categórica**, asignando **un color distinto a cada categoría**.  
Por ejemplo:
- `Setosa`
- `Versicolor`
- `Virginica`
Cuando se usa `hue`, Seaborn simplemente dibuja **una distribución independiente para cada categoría** y las superpone en el mismo gráfico.
### Limitación de `hue` en nuestro caso
El objetivo del último gráfico no era solo mostrar cada especie, sino **identificar explícitamente qué especies se solapan en cada intervalo del histograma**.
El problema es que `hue` **no puede representar combinaciones de categorías**.  
Solo puede asignar un color a **una categoría individual**, pero no puede generar automáticamente categorías como:
- `Setosa + Versicolor`
- `Setosa + Virginica`
- `Versicolor + Virginica`
- `Setosa + Versicolor + Virginica`
Estas combinaciones **no existen como valores reales en la columna `species`**, por lo que Seaborn no puede generarlas mediante `hue`.
### Qué se hizo en su lugar
Para poder mostrar esos solapamientos:
1. Se calcularon manualmente las **frecuencias por intervalo (`bin`) para cada especie**.
2. Después se identificaron las **zonas donde varias especies aparecen en el mismo intervalo**.
3. Finalmente se representaron esas combinaciones con **colores y tramas específicas**, añadiéndolas explícitamente a la leyenda.
De esta forma el gráfico no muestra solo las especies, sino también **las regiones donde sus distribuciones coinciden**, algo que `hue` por sí solo no puede representar.

## Ejercicio 7 - Visualización de la distribución por categorías con boxplot

El boxplot (o diagrama de caja) permite comparar cómo se distribuye una variable numérica según distintas categorías.

Vamos a comparar la distribución del largo del pétalo (petal_length) entre las distintas especies de flores del conjunto `df_iris`. La idea es que grafiques un boxplot que compare el largo del pétalo (`petal_length`) entre las diferentes especies (species).


1. Usa `seaborn.boxplot()` indicando el eje `x` con la especie y `y` con el largo del pétalo.
2. Es recomendable usar plt.figure() para ajustar el tamaño del gráfico si es necesario.

````python
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_length', data=df_iris)

plt.title('Análisis de Cuartiles: Longitud del Pétalo por Especie')
plt.show()
````

### Comentarios de Resolución
La pista indica usar `seaborn.boxplot()` con:

- `x` igual a la especie (`species`)
- `y` igual al largo del pétalo (`petal_length`)

Además, se recomienda usar `plt.figure()` para ajustar el tamaño del gráfico.
Este boxplot permite comparar la distribución de `petal_length` entre categorías.
Sintaxis importante:

- `x='species'` → variable categórica.
- `y='petal_length'` → variable numérica.
- `sns.boxplot(...)` → crea el diagrama de caja.
### Qué representa este boxplot
Este gráfico es un **boxplot (diagrama de caja y bigotes)** que muestra la **distribución de la variable `petal_length` para cada especie** del dataset. El eje X representa la **categoría (especie)** y el eje Y representa el **valor de la longitud del pétalo**.
Cada boxplot resume la distribución de los datos mediante cinco elementos principales:
- **Mediana (línea dentro de la caja)**  
  Representa el valor central de la distribución: el 50 % de los datos está por encima y el 50 % por debajo.
- **Caja (box)**  
  Representa el **rango intercuartílico (IQR)**, es decir, el rango donde se encuentra el **50 % central de los datos** (entre el primer cuartil Q1 y el tercer cuartil Q3).
- **Bigotes (whiskers)**  
  Muestran el rango de valores considerados normales dentro de la distribución.
- **Puntos aislados (outliers)**  
  Son observaciones que quedan fuera del rango esperado de los datos.
Este tipo de gráfico permite **comparar rápidamente la posición central, la dispersión y la presencia de valores atípicos entre diferentes categorías**.
### Interpretación aplicada a nuestros datos
En el gráfico se comparan las tres especies: **virginica, versicolor y setosa**.
Observando las **medianas**, las tres especies tienen valores similares de `petal_length`, situándose aproximadamente alrededor de **3.8–4 unidades**, lo que indica que el valor central de la longitud del pétalo es bastante parecido entre ellas.
Sin embargo, se pueden observar diferencias en la **dispersión de los datos**:
- **Setosa** muestra una distribución algo más amplia hacia valores altos, con datos que llegan aproximadamente hasta **7.4**.
- **Versicolor** presenta una dispersión similar, pero con un **valor atípico alto cercano a 7.7**.
- **Virginica** también presenta dispersión y muestra un **outlier bastante alto alrededor de 8.5**, lo que indica una observación con pétalos significativamente más largos que el resto.
En general, los rangos de las cajas se solapan bastante entre especies, lo que sugiere que **la variable `petal_length` por sí sola no separa completamente las especies en este dataset**, ya que sus distribuciones se mezclan en gran parte del rango de valores.

### Vamos a añadir algunas mejoras al gráfico
Este código mejora el gráfico añadiendo:
- nombres de ejes y título en español
- orden de especies según la mediana de `petal_length`
- colores pastel
- puntos individuales superpuestos
- línea de la media dentro de cada grupo
- rejilla horizontal suave para facilitar la lectura

### Qué añade cada mejora
- `palette='pastel'`  
  Aplica colores suaves para distinguir mejor cada especie.
- `order=orden`  
  Ordena las especies según la mediana de `petal_length`.
- `showmeans=True`  
  Muestra la **media** de cada grupo.
- `sns.stripplot(...)`  
  Añade los puntos individuales para ver la distribución real de las observaciones.
- `plt.grid(axis='y', alpha=0.25)`  
  Añade una rejilla horizontal suave para leer mejor los valores.

````python
import matplotlib.pyplot as plt
import seaborn as sns

df_box = df_iris.copy()

df_box['Especie'] = df_box['species'].map({
    'setosa': 'Setosa',
    'versicolor': 'Versicolor',
    'virginica': 'Virginica'
})

orden = (
    df_box.groupby('Especie')['petal_length']
    .median()
    .sort_values()
    .index
)

plt.figure(figsize=(9,6))

ax = sns.boxplot(
    data=df_box,
    x='Especie',
    y='petal_length',
    hue='Especie',    
    order=orden,
    palette='pastel',
    legend=False,
    showmeans=True,
    meanprops={
        'marker':'D',
        'markerfacecolor':'black',
        'markeredgecolor':'black',
        'markersize':6
    }
)

sns.stripplot(
    data=df_box,
    x='Especie',
    y='petal_length',
    order=orden,
    color='black',
    alpha=0.45,
    jitter=0.2,
    size=4
)

plt.title('Distribución de la longitud del pétalo por especie')
plt.xlabel('Especie')
plt.ylabel('Longitud del pétalo')
plt.grid(axis='y', alpha=0.25)

plt.show()
````

### Otra alternativa, solo boxplot más limpio , sin puntos

````python
plt.figure(figsize=(9, 6))
sns.boxplot(
    data=df_box,
    x='Especie',
    y='petal_length',
    hue='Especie',
    order=orden,
    palette='pastel',
    legend=False,
    showmeans=True
)
plt.title('Distribución de la longitud del pétalo por especie')
plt.xlabel('Especie')
plt.ylabel('Longitud del pétalo')
plt.grid(axis='y', alpha=0.25)
plt.show()
````

### Otra alternativa,`swarmplot` en vez de `stripplot` para evitar solapamiento de puntos

````python
plt.figure(figsize=(9, 6))
sns.boxplot(
    data=df_box,
    x='Especie',
    y='petal_length',
    hue='Especie',
    legend=False,
    order=orden,
    palette='pastel'
)
sns.swarmplot(
    data=df_box,
    x='Especie',
    y='petal_length',
    order=orden,
    color='black',
    size=4
)
plt.title('Distribución de la longitud del pétalo por especie')
plt.xlabel('Especie')
plt.ylabel('Longitud del pétalo')
plt.show()
````
